# [dbo].[verifyOtp]

```sql
/****** Object:  StoredProcedure [dbo].[verifyOtp]    Script Date: 7/13/2024 11:48:14 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE proc [dbo].[verifyOtp]
@Email varchar(50),
@otp numeric(4)
as 
begin
	if exists(select *from vf where Email=@Email)
	begin
		declare @dd1 datetime
		set @dd1=CONVERT(TIME(0),GETDATE())
		declare @i numeric(3)
		set @i=SUBSTRING(CONVERT(varchar(100),@dd1),16,2)

		declare @dd2 datetime 
		set @dd2=(select tm from vf where Email=@Email)
		declare @j numeric(3)
		set @j=SUBSTRING(CONVERT(varchar(100),@dd2),16,2)
		/*select @i
		select @j*/
		if(@i<=@j+5)
		begin
			declare @num numeric(4)
			set @num=(select otp from vf where Email=@Email)
			if (@num=@otp)
			begin
				delete from vf where Email=@Email
				
				select 1;
			end
			else
				/*if otp is wrong*/
				select 3;
		end
		else
			/*if time limit exceeded*/
			select 4;

	end
	else
		/*if user not found*/
		select 2;
end


GO
```
